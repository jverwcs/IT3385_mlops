```python
# === EDA ===
import os, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
os.makedirs("reports/figures", exist_ok=True)

df = pd.read_csv("datasets/datasets/03_Wheat_Seeds.csv")
print("Shape:", df.shape)
print(df.dtypes)
print("Missing values:\n", df.isna().sum())

expected = ['Area','Perimeter','Compactness','Length','Width','AsymmetryCoeff','Groove','Type']
print("Columns:", list(df.columns))
```

```python
# Class balance
ax = df['Type'].value_counts().sort_index().plot(kind='bar', title='Class Balance')
ax.set_xlabel('Type'); ax.set_ylabel('Count')
plt.tight_layout(); plt.savefig("reports/figures/class_balance.png"); plt.show()
```

```python
# Distributions
num_cols = ['Area','Perimeter','Compactness','Length','Width','AsymmetryCoeff','Groove']
for c in num_cols:
    plt.figure()
    sns.histplot(df[c], kde=True)
    plt.title(f'Distribution of {c}')
    plt.tight_layout(); plt.savefig(f"reports/figures/dist_{c}.png"); plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='Type', vars=num_cols, diag_kind='hist')
plt.savefig("reports/figures/pairplot.png"); plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlations")
plt.tight_layout(); plt.savefig("reports/figures/correlations.png"); plt.show()
```

```python
import pandas as pd
from pycaret.classification import *
import mlflow
import pycaret
```

```python
print("mlflow", mlflow.__version__, "| pycaret", pycaret.__version__)
```

```python
mlflow.set_tracking_uri("http://localhost:5000")
```

```python
mlflow.set_experiment("wheat_seeds_cls")
```

```python
from pycaret.classification import *

# PyCaret expects you to point to the target column; here it's 'Type'
s = setup(
    data=df,
    target='Type',
    session_id=123,
    normalize=True, 
    log_experiment=True,   
    experiment_name='wheat_seeds_cls',
    log_plots=True,               
    log_profile=False              
)
```

```python
best = compare_models()   # cross-validates many algorithms, returns the top one
best
```

```python
tuned_best = tune_model(best)
tuned_best
```

```python
final = finalize_model(tuned_best)
```

```python
plot_model(final, 'confusion_matrix')
```

```python
plot_model(final, 'class_report')
```

```python
plot_model(final, 'feature')
```
